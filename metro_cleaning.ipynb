```python
import pandas as pd
```

```python
metro = pd.read_csv('/Users/matteo/Downloads/Cities_with_metro.csv')
metro
```

```python
columns = ['Name', 'Service opened', 'Last expanded', 'Stations']
cols2 = ['System length', 'Annual ridership (millions)']
```

```python
for col in columns:
    temp = metro[col].str.split('[').tolist()
    temp_list = []
    for i in range(0, len(temp)):
        temp_list.append(temp[i][0])

    metro[col] = temp_list

for col in cols2:
    temp = metro[col].str.split(' ').tolist()
    temp_list = []
    for i in range(0, len(temp)):
        temp_list.append(temp[i][0])

    metro[col] = temp_list
```
